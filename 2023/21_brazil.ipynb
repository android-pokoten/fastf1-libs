```python
laps = session_race.laps.pick_driver('VER')[4:]
laps['LapTime']
```

```python
laps = laps.reset_index()
laps['LapTimeSec'] = laps['LapTime'].dt.total_seconds()
laps['LapTimeSec']
laps = laps.reset_index(drop=True)
print(laps['LapNumber'].count())
```

```python
avg_lap = laps['LapTimeSec'].sum() / laps['LapNumber'].count()
avg_lap
```

```python
laps['deltaLapTime'] = avg_lap - laps['LapTimeSec'] 
laps['deltaLapTime']
```

```python
laps['deltaLapTime'][:30]
#laps = laps[1:]
```

```python
laps['deltaCumsum'] = laps['deltaLapTime'].cumsum()
#laps['deltaCumsum'] = laps['deltaCumsum'] - avg_lap
laps['deltaCumsum']
```

```python
import matplotlib.pyplot as plt
import numpy as np

min_sec = laps['deltaCumsum'].min() - 3
max_sec = laps['deltaCumsum'].max() + 3

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(laps['LapNumber'], laps['deltaCumsum'])
#ax.plot(laps2['LapNumber'], laps2['deltaCumsum'])
ax.set_ylim([min_sec, max_sec])

plt.show() 
```
