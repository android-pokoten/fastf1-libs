```python
import fastf1
import fastf1.plotting
import seaborn as sns
import matplotlib.pyplot as plt

fastf1.plotting.setup_mpl(misc_mpl_mods=False)

driver_laps = session_race.laps.pick_driver('VER').pick_quicklaps().reset_index()
driver_laps2 = session_race.laps.pick_driver('HAM').pick_quicklaps().reset_index()

fig, ax = plt.subplots(figsize=(8, 8))

sns.scatterplot(data=driver_laps,
                x="LapNumber",
                y="LapTime",
                ax=ax,
                hue="Compound",
                palette=fastf1.plotting.COMPOUND_COLORS,
                s=80,
                linewidth=0,
                legend='auto')

sns.scatterplot(data=driver_laps2,
                x="LapNumber",
                y="LapTime",
                ax=ax,
                hue="Compound",
                palette=fastf1.plotting.COMPOUND_COLORS,
                s=80,
                linewidth=0,
                legend='auto')

ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")

ax.invert_yaxis()
plt.suptitle(f"{session_race.event['EventName']} {session_race.event.year} {'VER'} Race Laptime")

plt.grid(color='w', which='major', axis='both')
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
